```python
df=data_set.dropna()
```

```python
x=df.iloc[:,2:]
y=df.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import ElasticNet,Ridge,Lasso
from sklearn.metrics import mean_squared_error,r2_score
```

```python
m1=ElasticNet().fit(x_train,y_train)
train_score=m1.score(x_train,y_train)
pred=m1.predict(x_test)
test_score=r2_score(y_test,pred)


print(train_score)
print(test_score )
```

```python
alphas=[0.01,0.02,0.1,0.3,0.5,0.7,1]
```

```python
for a in alphas:
  model1=ElasticNet(alpha=a).fit(x,y)
  score=model1.score(x,y)
  pred=model1.predict(x)
  mse=mean_squared_error(y,pred)
  print('Alpha:{0:1f},R2:{1:1f},MSE:{2:1f},RMSE:{3:1F}'.format(a,score,mse,np.sqrt(mse)))
```

```python
m2=ElasticNet(alpha=0.01).fit(x_train,y_train)
train_score=m2.score(x_train,y_train)
pred=m2.predict(x_test)
test_score=r2_score(y_test,pred)
print(train_score)
print(test_score)
```

```python
m3=Ridge(alpha=0.01).fit(x_train,y_train)
train_score=m3.score(x_train,y_train)
pred=m3.predict(x_test)
test_score=r2_score(y_test,pred)
print(train_score)
print(test_score)
```

```python
m4=Lasso(alpha=0.01).fit(x_train,y_train)
train_score=m4.score(x_train,y_train)
pred=m4.predict(x_test)
test_score=r2_score(y_test,pred)
print(train_score)
print(test_score)
```
