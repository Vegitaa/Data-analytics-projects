```python
def removeouliers(col): # removing htew outlier using IQR test (Inter quantile range test )
  quant1,quant2=col.quantile([0.25,0.75])
  IQR=quant2-quant1
  lower_range=quant1-(1.5*IQR)
  upper_range=quant2+(1.5*IQR)
  return lower_range,upper_range

lowscore,highscore=removeouliers(df['Family'])
df['Family']=np.where(df['Family']>highscore,highscore,df['Family'])
df['Family']=np.where(df['Family']<lowscore,lowscore,df['Family'])
```
